```python
# FunÃ§Ã£o calcula mÃ©dia de uma coluna de um dataframe

def calcula_media(coluna):

    media = coluna.mean()
    return media

print("MÃ©dia da temperatura do Ar", calcula_media(dados['Temperatura Ar [K]']))
print("MÃ©dia da temperatura do Processo", calcula_media(dados['Temperatura Processo [K]']))
print("MÃ©dia da velocidade rotaÃ§Ã£o", calcula_media(dados['Velocidade Rotacao [rpm]']))
```

```python
# FunÃ§Ã£o soma de uma coluna de um dataframe

def soma_coluna(coluna):

    soma = coluna.sum()
    return soma

print("Soma da temperatura do Ar", soma_coluna(dados['Temperatura Ar [K]']))
print("Soma da temperatura do Processo", soma_coluna(dados['Temperatura Processo [K]']))
print("Soma da velocidade de rotaÃ§Ã£o", soma_coluna(dados['Velocidade Rotacao [rpm]']))
```

```python
# ExercÃ­cio 

# Mostre como calcular a diferenca entre as colunas velocidade de rotaÃ§Ã£o e torque. 
# Utilize estrutura com funÃ§Ã£o def.

def calcula_diferenca(coluna1, coluna2):

    diferenca = coluna1 - coluna2

    return diferenca

calcula_diferenca(dados['Velocidade Rotacao [rpm]'], dados['Torque [Nm]'])
```
