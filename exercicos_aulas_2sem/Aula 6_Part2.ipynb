```python
# Função calcula média de uma coluna de um dataframe

def calcula_media(coluna):

    media = coluna.mean()
    return media

print("Média da temperatura do Ar", calcula_media(dados['Temperatura Ar [K]']))
print("Média da temperatura do Processo", calcula_media(dados['Temperatura Processo [K]']))
print("Média da velocidade rotação", calcula_media(dados['Velocidade Rotacao [rpm]']))
```

```python
# Função soma de uma coluna de um dataframe

def soma_coluna(coluna):

    soma = coluna.sum()
    return soma

print("Soma da temperatura do Ar", soma_coluna(dados['Temperatura Ar [K]']))
print("Soma da temperatura do Processo", soma_coluna(dados['Temperatura Processo [K]']))
print("Soma da velocidade de rotação", soma_coluna(dados['Velocidade Rotacao [rpm]']))
```

```python
# Exercício 

# Mostre como calcular a diferenca entre as colunas velocidade de rotação e torque. 
# Utilize estrutura com função def.

def calcula_diferenca(coluna1, coluna2):

    diferenca = coluna1 - coluna2

    return diferenca

calcula_diferenca(dados['Velocidade Rotacao [rpm]'], dados['Torque [Nm]'])
```
